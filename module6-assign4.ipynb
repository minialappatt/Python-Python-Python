```python
import pandas as pd
import matplotlib.pyplot as plt

# Importing the CSV file
data = pd.read_csv('cars.csv')

# Extracting the 'cyl' and 'model' columns
cyl_counts = data['cyl'].value_counts()
model_counts = data['model'].value_counts()
```

```python
# Creating the pie chart for 'cyl'
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.pie(cyl_counts, labels=cyl_counts.index, autopct='%1.1f%%')
plt.title('Cylinder Distribution')
```

```python
# Creating the pie chart for 'model'
#plt.subplot(1, 2, 2)
plt.pie(model_counts, labels=model_counts.index, autopct='%1.1f%%')
plt.title('Model Distribution')
# Adjusting layout and displaying the chart
plt.tight_layout()
plt.show()
```
