1. Load iris data from load_iris function from sklearn.datasets package
2. From the dataset extract the data property
3. Train an Agglomerative Clustering model based on the data
4. Plot dendrogram to visualize the clustering linkage

```python
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Load iris data
iris = load_iris()

# Extract the data property
X = iris.data

# Train Agglomerative Clustering model
model = AgglomerativeClustering(n_clusters=3)  # Specify the number of clusters
model.fit(X)

# Plot dendrogram
linkage_matrix = linkage(X, method='ward')  # Perform hierarchical clustering using Ward's method
dendrogram(linkage_matrix)
plt.title('Dendrogram')
plt.xlabel('Data points')
plt.ylabel('Distance')
plt.show()
```
