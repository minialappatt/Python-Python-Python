```python
import pandas as pd

# Read the CSV file and store it in a DataFrame
churn_df = pd.read_csv('customer_churn.csv')

# Display the DataFrame
print(churn_df)
```

#1. Now from the churn DataFrame, try to sort the data by the tenure column
according to the descending order.

```python
sorted_churn = churn_df.sort_values(by='tenure', ascending=False)
print(sorted_churn)
```

#2. Fetch all the records that are satisfying the following condition:
a. Tenure>50 and the gender as ‘Female’
b. Gender as ‘Male’ and SeniorCitizen as 0
c. TechSupport as ‘Yes’ and Churn as ‘No’
d. Contract type as ‘Month-to-month’ and Churn as ‘Yes

```python
#a
condition_a = churn_df[(churn_df['tenure'] > 50) & (churn_df['gender'] == 'Female')]
print(condition_a)
```

```python
#b
condition_b = churn_df[(churn_df['gender'] == 'Male') & (churn_df['SeniorCitizen'] == 0)]
print(condition_b)
```

```python
#c
condition_c = churn_df[(churn_df['TechSupport'] == 'Yes') & (churn_df['Churn'] == 'No')]
print(condition_c)
```

```python
#d
condition_d = churn_df[(churn_df['Contract'] == 'Month-to-month') & (churn_df['Churn'] == 'Yes')]
print(condition_d)
```
