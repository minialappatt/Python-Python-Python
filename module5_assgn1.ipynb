9. Fetch all the records that are satisfying the following condition:
a. Tenure>50 and the gender as ‘Female’.
b. Gender as ‘Male’ and SeniorCitizen as 0.
c. TechSupport as ‘Yes’ and Churn as ‘No’.
d. Contract type as ‘Month-to-month’ and Churn as ‘Yes’.

```python
condition_a = churn_df[(churn_df['tenure'] > 50) & (churn_df['gender'] == 'Female')]
print(condition_a)
condition_b = churn_df[(churn_df['gender'] == 'Male') & (churn_df['SeniorCitizen'] == 0)]
print(condition_b)
condition_c = churn_df[(churn_df['TechSupport'] == 'Yes') & (churn_df['Churn'] == 'No')]
print(condition_c)
condition_d = churn_df[(churn_df['Contract'] == 'Month-to-month') & (churn_df['Churn'] == 'Yes')]
print(condition_d)
```

10. Use a for loop to calculate the number of customers that are getting the tech support and are male senior citizen.

```python
count = 0

for index, row in churn_df.iterrows():
    if row['TechSupport'] == 'Yes' and row['gender'] == 'Male' and row['SeniorCitizen'] == 1:
        count += 1

print("Number of customers getting tech support and are male senior citizens:", count)
```
