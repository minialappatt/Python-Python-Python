```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
xpoints = np.array([0,5])
ypoints = np.array([0,200])
```

```python
plt.plot(xpoints,ypoints)
plt.show()
```

```python
plt.plot(xpoints,ypoints,'o')
plt.show()
```

```python
plt.plot(xpoints,ypoints,marker='*')
plt.show()
```

```python
plt.plot(xpoints,ypoints,marker='D')
plt.show()
```

```python
xpoints = np.array([1,2,4,6])
ypoints = np.array([3,7,4,10])
plt.plot(xpoints,ypoints,'o')
plt.show()
```

```python
xpoints = np.array([1,2,4,6])
ypoints = np.array([3,7,4,10])
plt.plot(xpoints,ypoints, linestyle='dotted')
plt.show()
```

```python
xpoints = np.array([1,2,4,6])
ypoints = np.array([3,7,4,10])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("This is demo graph")
plt.plot(xpoints,ypoints, linestyle='dotted')
plt.show()
```

```python
font1 = {'family':'serif','color':'red','size': 20}
font2 = {'family':'serif', 'color': 'blue', 'size': 10}

plt.title("This is title", fontdict=font1)
plt.xlabel("This is x-axis", fontdict=font2)
plt.ylabel("This is y-axis", fontdict=font2)

plt.plot(xpoints,ypoints)
plt.show()
```

```python
bikes = pd.read_csv("bike_shairing_hourly.csv")
```

```python
bikes.head()
```

```python
plt.plot(bikes.season)
```

```python
plt.plot(bikes.temp)
```

```python
plt.plot(bikes.cnt)
```

```python
plt.plot(bikes.casual)
```

```python
plt.plot(bikes.season,bikes.cnt)
```

```python
df_by_seasons = bikes.groupby("season")
```

```python
df_seasons = df_by_seasons.cnt.sum().sort_values()
```

```python
df_seasons
```

```python
plt.plot(bikes.season.unique(), df_by_seasons.cnt.sum())
```

```python
plt.ticklabel_format(style='plain')
```

```python
plt.ticklabel_format(style='plain')
plt.plot(bikes.season.unique(), df_by_seasons.cnt.sum())
```

```python
fig, ax = plt.subplots()

index = bikes.season.unique()

bar_width = 0.25

values = df_by_seasons.cnt.sum()

ax.bar(index,values, bar_width)
```

```python
fig, ax = plt.subplots()
ax.scatter(bikes.casual, bikes.registered)
plt.show()
```

```python
fig, ax = plt.subplots()

values = df_by_seasons.cnt.sum()
labels = ["s1" , "s2", "s3", "s4"]
ax.pie(values,labels=labels)
```
